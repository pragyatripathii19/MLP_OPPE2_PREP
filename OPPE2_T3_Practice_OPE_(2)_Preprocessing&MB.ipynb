## **Common Instructions for Question 9 and 10**
## **Rules for Imputing the missing(NaN) or Unknown values:**
- Calculating statistical values (such as mean, median, mode) for each column in the training dataset.
- Applying these calculated statistical values to replace missing (NaN) and unknown values in both the training and test datasets.
- Ensure that the calculation of statistical values excludes any rows containing missing or unknown values.

- Replace Unknown values in the **"Age"** feature with the Mean value in that.
- Replace Unknown values in the **"Location"** feature with the constant value **"Other"**.
- Replace Unknown values in the **"GameDifficulty"** feature with the Most Frequent value in that.
- Replace Unknown values in the **"InGamePurchases"** feature with the constant value '0'.


- Write the answers related to the above imputation in below questions respectively.


```python
X_train.info()
```

```python
gd_mode=X_train["GameDifficulty"].mode()
print(gd_mode)

print('--------------------------------')

gd_mode_tr=X_train["GameDifficulty"].mode()[0]
print(gd_mode_tr)
```

```python
#col 0 has no missing values so all good

#col 1 "age" missing values to be imputed with mean
age_mean=X_train['Age'].mean()
X_train['Age']=X_train['Age'].fillna(age_mean)
X_test['Age']=X_test['Age'].fillna(age_mean)

#col 2  has no missing values

#col 3 "Location" has to be imputed with "Other" for its NA values
loc_val="Other"
X_train['Location']=X_train['Location'].fillna(loc_val)
X_test['Location']=X_test['Location'].fillna(loc_val)

#col 4 and 5 have no missing values

#col 6
X_train['InGamePurchases']=X_train['InGamePurchases'].fillna(0)
X_test['InGamePurchases']=X_test['InGamePurchases'].fillna(0)

#col 7 to be imputed with most frequent value
gd_mode=X_train["GameDifficulty"].mode()[0]
X_train["GameDifficulty"]=X_train["GameDifficulty"].fillna(gd_mode)
X_test["GameDifficulty"]=X_test["GameDifficulty"].fillna(gd_mode)

#col 8-11 are good too
```

```python
from sklearn.ensemble import RandomForestClassifier
?RandomForestClassifier
```

```python
from scipy.stats import uniform
?uniform
```

```python
rf=RandomForestClassifier(random_state=42)

param_distributions={'n_estimators': list(range(2,100)),
                     'max_depth': list(range(1,11)),
                     'min_impurity_decrease': uniform(loc=0,scale=5)}


model=RandomizedSearchCV(estimator=rf,
                         param_distributions=param_distributions,
                         random_state = 42,
                         n_iter=5,
                         cv=3,
                         n_jobs= -1,
                         verbose=2)

model.fit(X_train, y_train)
print(model.best_params_['n_estimators']) #ans is 16
```
